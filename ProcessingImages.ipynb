# Using Matplotlib.image Library to Read & Print the image, :)

```python
# Importng Image Module from matplotlib.image

import matplotlib.image as mpimg
import matplotlib.pyplot as plt
```

```python
# Loading & Storing our image in the "img" variable

img = mpimg.imread('/content/HD-wallpaper-cool-vampire-art.jpg') # It stores images as numpy.ndarrays
```

```python
print(img.shape) # Prints out image's height & width along with the channel of image (1 - Grayscale, 3 - RGB)
```

```python
print(img) # Prints the image in n-diamensional numpy array
```

```python
# Displaying the image in pictorial format

img_plot = plt.imshow(img)
plt.show()
```

# The image above is 800 x 450, and is RGB. We can use the Pillow Library to Resize it accordingly.

```python
from PIL import Image   # Importing Image module from Pillow (PIL)
img = Image.open('/content/HD-wallpaper-cool-vampire-art.jpg')  # Loading the Image
img_resized = img.resize((720, 480))  # Applying the resize method on our image, resizing it to 480p
img_resized.save('updatedVamp.jpg')   # Saving our resized image
```

```python
# Displaying the resized image

img_resized = mpimg.imread('/content/updatedVamp.jpg')
img_resized_plot = plt.imshow(img_resized)
plt.show()
```

```python
# Checking the dimensions & channel of the image
print(img_resized.shape)
```

# Time to convert it into a Grayscale Image, using the OpenCV library.

```python
# Import the OpenCV library

import cv2

img = cv2.imread('/content/HD-wallpaper-cool-vampire-art.jpg') # Using the original Image (not the resized one) & loading it
grayscale_image = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY) # Applying the convertColor method on our image to convert it into grayscale
```

```python
# Checking the dimensions & especially the CHANNEL if it changed from 3 (RGB) to 1 (GrayScale)

print(grayscale_image.shape)
```

* As we can see, the third parameter is gone and is by default set the "1" thus not necessary to show it.

```python
# Importing the show method for OpenCV2 using colab patches

from google.colab.patches import cv2_imshow

# Displaying & saving the final result

cv2_imshow(grayscale_image)
cv2.imwrite('updatedGrayscaleVamp.jpg', grayscale_image)
```
